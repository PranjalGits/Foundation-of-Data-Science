# **Presentation of data by tables**

Presenting data in tables is a common way to organize and display information in a structured format. Tables allow for easy comparison, analysis, and understanding of data.

# 1) Using Pandas DataFrame
Pandas is a powerful library in Python for data manipulation and analysis. It provides a DataFrame data structure that is well-suited for organizing and presenting data in tabular form.

```python
import pandas as pd

# Sample data
data = {
    'Name': ['Nancy', 'Anuj', 'Shivtar', 'Prabal', 'Anya'],
    'Age': [25, 30, 22, 35, 28],
    'Gender': ['Female', 'Male', 'Male', 'Male', 'Female'],
    'City': ['Bhopal', 'Indore', 'Ujjain', 'Pune', 'Delhi']
}

# Create a DataFrame
df = pd.DataFrame(data)

# Display the DataFrame as a table
print("DataFrame:")
print(df)

# Alternatively, you can display specific rows or columns
print("\nSelected Columns:")
print(df[['Name', 'Age']])

# You can also filter rows based on conditions
print("\nFiltered Rows (Age > 25):")
print(df[df['Age'] > 25])
```

# 2) Using PrettyTable
If you prefer a more visually appealing table, you can use the PrettyTable library.

```python
from prettytable import PrettyTable

# Create a new PrettyTable instance
table = PrettyTable()

# Define table columns
table.field_names = ["Name", "Age", "Gender", "City"]

# Add rows to the table
table.add_row(["Mohit", 25, "Male", "Vidisha"])
table.add_row(["Gori", 30, "Female", "Ujjain"])
table.add_row(["Pranjal", 35, "Male", "USA"])
table.add_row(["Ujala", 28, "Female", "Bhopal"])

# Print the table
print(table)
```

# 3) Using Tabulate
Tabulate is another Python library that allows you to generate formatted tables easily.

```python
from tabulate import tabulate

# Define data as a list of lists
data = [
    ["Mohit", 25, "Male", "New York"],
    ["Anya", 30, "Female", "Los Angeles"],
    ["Pranjal", 35, "Male", "Chicago"],
    ["Tanu", 28, "Female", "Houston"]
]

# Define column headers
headers = ["Name", "Age", "Gender", "City"]

# Print the tabulated data
print(tabulate(data, headers=headers, tablefmt="grid"))
```

This will produce a grid-style table similar to PrettyTable.

# **construction of frequency distributions for discrete and continuous data**

Constructing frequency distributions for discrete and continuous data involves organizing data into intervals or categories and counting the frequency of values within each interval or category.

# 1)  Discrete Data
For discrete data, which consists of distinct and separate values, we can create a frequency distribution by counting the occurrences of each unique value.

```python
import pandas as pd

# Sample discrete data
data_discrete = [2, 3, 1, 5, 2, 4, 3, 2, 1, 5, 4, 3, 2, 1, 4, 5, 3, 2, 4, 1]

# Create a DataFrame
df_discrete = pd.DataFrame(data_discrete, columns=['Values'])

# Calculate frequency distribution using value_counts() function
frequency_discrete = df_discrete['Values'].value_counts()

# Display frequency distribution for discrete data
print("Frequency Distribution for Discrete Data:")
print(frequency_discrete)
```

# 2) Continuous Data
For continuous data, we often need to create intervals (bins) to group the data. We can use histograms to visualize the frequency distribution of continuous data.

```python
import matplotlib.pyplot as plt
import numpy as np

# Sample continuous data
data_continuous = np.random.normal(loc=50, scale=10, size=100)

# Create histogram to visualize frequency distribution
plt.figure(figsize=(8, 8))
plt.hist(data_continuous, bins=10, color='green', edgecolor='black', alpha=0.7)
plt.title('Histogram: Frequency Distribution for Continuous Data')
plt.xlabel('Values')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

In this code:

1)We generate sample continuous data using NumPy's random.normal() function to create a normal distribution.

2)We create a histogram using Matplotlib to visualize the frequency distribution of the continuous data.

3)The data is divided into bins (intervals), and the height of each bar represents the frequency of values within that bin.

# **Graphical representation of a frequency distribution by histogram and frequency polygon, cumulative frequency distributions.**

```python
import numpy as np
import matplotlib.pyplot as plt

# Sample data
data = np.random.randint(1,11, size=50)  # Generate random integers between 1 and 10

# Create a histogram
plt.figure(figsize=(10, 5))
plt.hist(data, bins=np.arange(1, 12) - 0.5, color='skyblue', edgecolor='black', alpha=0.7)
plt.title('Histogram: Frequency Distribution')
plt.xlabel('Values')
plt.ylabel('Frequency')
plt.xticks(range(1, 11))
plt.grid(True)
plt.show()

# Create a frequency polygon
plt.figure(figsize=(10, 5))
plt.hist(data, bins=np.arange(1, 12) - 0.5, color='lightgreen', edgecolor='black', alpha=0.7)
plt.plot(np.arange(1, 11), np.histogram(data, bins=np.arange(1, 12))[0], marker='o', color='red', linestyle='--')
plt.title('Frequency Polygon: Frequency Distribution')
plt.xlabel('Values')
plt.ylabel('Frequency')
plt.xticks(range(1, 11))
plt.grid(True)
plt.show()

# Create a cumulative frequency distribution
plt.figure(figsize=(10, 5))
plt.hist(data, bins=np.arange(1, 16) - 0.5, color='orange', edgecolor='black', alpha=0.7, cumulative=True)
plt.title('Cumulative Frequency Distribution')
plt.xlabel('Values')
plt.ylabel('Cumulative Frequency')
plt.xticks(range(1, 11))
plt.grid(True)
plt.show()
```

We create a cumulative frequency distribution by setting the cumulative parameter to True in the histogram function.

# 1) Histogram
A histogram is a graphical representation of the frequency distribution of continuous data. It consists of a series of adjacent rectangles (bars), where the width of each bar represents the interval or bin, and the height represents the frequency of observations within that interval

```python
import numpy as np
import matplotlib.pyplot as plt

# Sample data
data = np.random.randint(1, 11, size=50)  # Generate random integers between 1 and 10

# Create a histogram
plt.figure(figsize=(10, 5))
plt.hist(data, color='skyblue', edgecolor='black', alpha=0.7)
plt.title('Histogram: Frequency Distribution')
plt.xlabel('Values')
plt.ylabel('Frequency')
plt.xticks(range(1, 11))
plt.grid(True)
plt.show()
```

# 2) Frequency Polygon
A frequency polygon is a graphical representation of the frequency distribution of data points connected by line segments. It is similar to a histogram but uses lines instead of bars to represent the frequencies.

```python
import numpy as np

# Define the data
ages = [22, 25, 30, 28, 35, 32, 40, 38, 42, 45, 50, 55, 60, 65, 70]

# Create a frequency distribution
hist, bins = np.histogram(ages, bins=5)

# Calculate the midpoints of the bins
midpoints = [(bins[i] + bins[i+1]) / 2 for i in range(len(bins)-1)]

# Plot the frequency polygon
plt.plot(midpoints, hist, marker='o', linestyle='-')
plt.title('Frequency Polygon of Ages')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

# 3) Cumulative Frequency Distribution
A cumulative frequency distribution represents the cumulative sum of frequencies up to a certain point. It shows the total number of observations up to each interval or value.

```python
# Calculate the cumulative sum of frequencies
cumulative_freq = np.cumsum(hist)

# Plot the cumulative frequency distribution
plt.plot(midpoints, cumulative_freq, marker='o', linestyle='-')
plt.title('Cumulative Frequency Distribution of Ages')
plt.xlabel('Age')
plt.ylabel('Cumulative Frequency')
plt.grid(True)
plt.show()
```

#**Data Pre-processing- Knowing Data, Data Cleaning, Data Integration, Data Selection, Data Transformation**

Data preprocessing is a crucial step in the data analysis pipeline. It involves several operations aimed at preparing raw data for analysis and modeling. key steps involved in data preprocessing:

### 1. Knowing Data:

Understanding the characteristics of your dataset is essential before proceeding with any analysis. This step involves:

- Exploratory Data Analysis (EDA): Analyzing the structure, patterns, and relationships within the dataset using statistical methods and visualization techniques.
- Understanding Data Types: Identifying the types of data (e.g., numerical, categorical, ordinal) present in the dataset.
- Data Summary: Summarizing basic statistics such as mean, median, standard deviation, and quartiles for numerical features, and frequency distributions for categorical features.

### 2. Data Cleaning:

Data cleaning aims to identify and handle missing values, outliers, and inconsistencies in the dataset. Key tasks include:

- Handling Missing Values: Imputing missing values using techniques like mean/median imputation, interpolation, or deletion.
- Handling Outliers: Identifying and treating outliers using methods such as trimming, winsorization, or transformation.
- Handling Inconsistencies: Addressing inconsistencies in data formats, units, and representations.

### 3. Data Integration:

Data integration involves combining data from multiple sources to create a unified dataset. This step includes:

- Data Consolidation: Merging datasets with shared identifiers or keys.
- Handling Redundancy: Removing duplicate records and resolving conflicts between datasets.

### 4. Data Selection:

Data selection involves selecting relevant features or variables for analysis while discarding irrelevant or redundant ones. Key considerations include:

- Feature Relevance: Identifying features that contribute meaningfully to the analysis and modeling objectives.
- Dimensionality Reduction: Using techniques like Principal Component Analysis (PCA) or feature selection algorithms to reduce the number of features.

### 5. Data Transformation:

Data transformation involves converting the dataset into a format suitable for analysis and modeling. This step includes:

- Normalization/Standardization: Scaling numerical features to a standard range to prevent biases in modeling algorithms.
- Encoding Categorical Variables: Converting categorical variables into numerical representations using techniques like one-hot encoding or label encoding.
- Feature Engineering: Creating new features from existing ones or transforming features to enhance predictive power.

By performing these preprocessing steps, you can ensure that the data is clean, consistent, and structured, making it suitable for analysis, modeling, and interpretation. Effective data preprocessing can significantly improve the accuracy and reliability of analytical results.

```python
dataset = [10,12,13,14,15,16,17,18,19,105,12,13,5,16,-15]
```

The Z-score, also known as standard score or standard deviation score, is a statistical measurement that describes a value's relationship to the mean of a group of values. It indicates how many standard deviations a data point is from the mean of the dataset.

The formula to calculate the Z-score of a data point

X is:

Z = X−μ/ σ

Where:


X is the individual data point,

μ is the mean of the dataset,

σ is the standard deviation of the dataset

```python
import numpy as np
outliers=[]

def detect_outliers(Data):
  threshold=3  # 3rd Standard Deviation
  mean=np.mean(Data)
  std=np.std(Data)

  for i in Data:
    z_score = (i-mean)/std
    if np.abs(z_score)>threshold:      ## abs-absolute function convert the negative value to positive value like -5 to 5
      outliers.append(i)

  return outliers
```

```python
detect_outliers(dataset)
```
