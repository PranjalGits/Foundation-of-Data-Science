**1. Write a Python program to draw a line using given axis values taken
from a text file, with suitable label in the x axis, y axis and a title.
Test Data:
test.txt
1 2

3 1**

```python
import matplotlib.pyplot as plt

# Define the data points
x = [1, 2, 3]
y = [2, 4, 1]

# Plot the data
plt.plot(x, y)

# Label the axes and title
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Line Plot')

# Display the plot
plt.show()
```

**2. Write a Python program to draw line charts of the financial data of
Alphabet Inc. between October 3, 2016 to October 7, 2016.
Sample Financial data (fdata.csv):
Date, Open, High, Low, Close
10-03-16,774.25,776.065002,769.5,772.559998
10-04-16,776.030029,778.710022,772.890015,776.429993
10-05-16,779.309998,782.070007,775.650024,776.469971
10-06-16,779,780.47998,775.539978,776.859985
10-07-16,779.659973,779.659973,770.75,775.080017**

```python
import matplotlib.pyplot as plt
import pandas as pd

# Sample Financial data
data = {
    'Date': ['10-03-16', '10-04-16', '10-05-16', '10-06-16', '10-07-16'],
    'Open': [774.25, 776.030029, 779.309998, 779, 779.659973],
    'High': [776.065002, 778.710022, 782.070007, 780.47998, 779.659973],
    'Low': [769.5, 772.890015, 775.650024, 775.539978, 770.75],
    'Close': [772.559998, 776.429993, 776.469971, 776.859985, 775.080017]
}

# Create DataFrame
df = pd.DataFrame(data)
df['Date'] = pd.to_datetime(df['Date'])

# Plotting
plt.figure(figsize=(10, 6))
plt.plot(df['Date'], df['Open'], marker='o', label='Open')
plt.plot(df['Date'], df['High'], marker='o', label='High')
plt.plot(df['Date'], df['Low'], marker='o', label='Low')
plt.plot(df['Date'], df['Close'], marker='o', label='Close')

# Formatting
plt.title('Alphabet Inc. Stock Prices (Oct 3 - Oct 7, 2016)')
plt.xlabel('Date')
plt.ylabel('Price')
plt.xticks(rotation=45)
plt.grid(True)
plt.legend()

# Show plot
plt.tight_layout()
plt.show()
```

**3. Write a Python program to plot two or more lines with different styles.
The code snippet gives the output shown in the following screenshot:**

```python
import matplotlib.pyplot as plt

# Data for lines
x = [1, 2, 3, 4, 5]
y1 = [1, 2, 3, 4, 5]
y2 = [5, 4, 3, 2, 1]

# Plotting lines with different styles
plt.plot(x, y1, linestyle='--', label='Line 1')
plt.plot(x, y2, linestyle=':', label='Line 2')

# Labeling and display
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Plot with Different Line Styles')
plt.legend()
plt.grid(True)  # Enable grid
plt.show()
```
