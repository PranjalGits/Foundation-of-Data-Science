**8.  Write a Pandas program to join the two given dataframes along rows and
assign all data.**

**Test Data:
student_data1:
student_id name marks
0 S1 Danniella Fenton 200
1 S2 Ryder Storey 210
2 S3 Bryce Jensen 190
3 S4 Ed Bernal 222
4 S5 Kwame Morin 199
student_data2:
student_id name marks
0 S4 Scarlette Fisher 201
1 S5 Carla Williamson 200
2 S6 Dante Morse 198
3 S7 Kaiser William 219
4 S8 Madeeha Preston 201**

```python

# Test Data
student_data1 = {
    'student_id': ['S1', 'S2', 'S3', 'S4', 'S5'],
    'name': ['Danniella Fenton', 'Ryder Storey', 'Bryce Jensen', 'Ed Bernal', 'Kwame Morin'],
    'marks': [200, 210, 190, 222, 199]
}

student_data2 = {
    'student_id': ['S4', 'S5', 'S6', 'S7', 'S8'],
    'name': ['Scarlette Fisher', 'Carla Williamson', 'Dante Morse', 'Kaiser William', 'Madeeha Preston'],
    'marks': [201, 200, 198, 219, 201]
}

# Create DataFrames
df1 = pd.DataFrame(student_data1)
df2 = pd.DataFrame(student_data2)

# Join the two dataframes along rows and assign all data
joined_df = pd.concat([df1, df2], ignore_index=True)

# Display the joined dataframe
print(joined_df)
```

**9. Write a Pandas program to join the two given dataframes along columns

**10. Write a Pandas program to append rows to an existing DataFrame and
display the combined data.**

```python

# Existing DataFrame
student_data1 = {
    'student_id': ['S1', 'S2', 'S3', 'S4', 'S5'],
    'name': ['Danniella Fenton', 'Ryder Storey', 'Bryce Jensen', 'Ed Bernal', 'Kwame Morin'],
    'marks': [200, 210, 190, 222, 199]
}
df1 = pd.DataFrame(student_data1)

# New Row(s) to append
new_row = pd.Series(['S6', 'Scarlette Fisher', 205], index=['student_id', 'name', 'marks'])

# Append the new row(s) to the existing DataFrame
df1 = df1.append(new_row, ignore_index=True)

# Display the combined data
print(df1)
```
